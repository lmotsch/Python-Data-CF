### Task 6: The team now wants to target different types of spenders in their marketing campaigns. 
1) If the mean of the prices of products purchased by a user is lower than 10, then flag them as a Low spender.
2) If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a High spender.

```python
# Step 1: First create a variable to store the mean of 'prices' grouped by 'user_id'
ords_prods_merge['avg_spend_user'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Step 2: Let's check the output
ords_prods_merge[['user_id', 'prices', 'avg_spend_user']].head(25)
```

#### Deriving columns with loc()

```python
ords_prods_merge.loc[ords_prods_merge['avg_spend_user'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_spend_user'] >= 10, 'spending_flag'] = 'High spender'
```

#### Checking frequency and sample

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'avg_spend_user', 'spending_flag']].head(25)
```

### Task 7: In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers.
If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”

If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”

If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# Step 1: First create a variable to store the median of 'days_since_prior_order' grouped by 'user_id'
ords_prods_merge['last_order_median'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Step 2: Let's check the output
ords_prods_merge[['user_id', 'days_since_prior_order', 'last_order_median']].head(25)
```

Integers and stable per user -> seems to be right

#### Deriving columns with loc()

```python
ords_prods_merge.loc[ords_prods_merge['last_order_median'] > 20, 'active_customer_status'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['last_order_median'] > 10) & (ords_prods_merge['last_order_median'] <= 20), 'active_customer_status'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['last_order_median'] <= 10, 'active_customer_status'] = 'Frequent customer'
```

#### Checking frequency and sample

```python
ords_prods_merge['active_customer_status'].value_counts(dropna = False)
```

```python
ords_prods_merge[ords_prods_merge['active_customer_status'].isna()]
```

NaN only shows one customer who only ordered one time - so it would make sense as there is no repeat purchase.

```python
users_one_order = (
    ords_prods_merge.groupby('user_id')['order_number']
    .max()                               # find the max order per user
    .reset_index()                       # back to DataFrame
    .query("order_number == 1")          # keep only max = 1
)

print(users_one_order)
```

Confirmed that there is only one user where this applies

### Creating Line Charts & sampling data

```python
#First we set the seed, a random but reproducible pattern
np.random.seed(4)
# Create 32,404,859 random numbers (done using the len() function)
# Check if value in a row is less than or equal to 0.7. That's our cutoff for the sample.
# Create a new list "dev" which features the results of the test per row (boolean).
dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
#Split dataframe into two samples with the help of our list
#Store 70% in our big list
big = ords_prods_merge[dev]
#Store 30% in our small list
small = ords_prods_merge[~dev]
```

```python
#Check length
len(ords_prods_merge)
```

```python
len(big) + len(small)
```

```python
#reduce sample to necessary column only
df_2 = small[['orders_day_of_week','prices']]
```

```python
#create line chart
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

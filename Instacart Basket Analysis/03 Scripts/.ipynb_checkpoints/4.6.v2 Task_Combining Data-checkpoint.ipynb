We can combine them using the product_id column

```python
#Create a view of the merged tables to check for issues. Cant merge the two dataframes as the previously combined one already has a '_merge' column. Need to drop the old merge column.
df_orders_products_combined = df_orders_products_combined.drop(['_merge'], axis=1)
```

### Mangling the data

```python
# Step 1: Dropping eval_set column, as it contains unnecessary information for our purpose
df_ords = df_ords.drop(columns = ['eval_set'])
```

```python
# Step 2: Checking for different variables and their frequencies
df_ords['days_since_prior_order'].value_counts(dropna = False)
```

```python
# Step 3: Renaming of a column
df_ords.rename(columns = {'order_dow' : 'orders_day_of_week'}, inplace = True)
```

```python
df_ords.head()
```

```python
# Step 4: Changing the data type of order_id
df_ords['order_id'] = df_ords['order_id'].astype('str')
```

```python
df_ords['order_id'].dtype
```

### Transforming the dataframe:

```python
# Step 1: We load the department info as a dataframe
df_dep = pd.read_csv(os.path.join(path, '02 Data', 'Original Data', 'departments.csv'), index_col = False)
```

```python
df_dep.head()
```

```python
# Step 2: Transposing the dataframe
df_dep.T
```

```python
# Step 3: We save the df as a copy, bc the previously seen table is only a visualization
df_dep_t = df_dep.T
```

```python
df_dep_t
```

```python
# Step 4: Reset the index of the table
df_dep_t.reset_index()
```

```python
# Step 5: Take first row of df_dep_t as header
new_header = df_dep_t.iloc[0]
```

```python
# Step 6: Test new header
new_header
```

```python
# Step 7: We create a copy of the df, omitting the first row
df_dep_t_new = df_dep_t[1:]
```

```python
df_dep_t_new.head()
```

```python
# Step 8: Replace the column names with the new header names
df_dep_t_new.columns = new_header
```

```python
df_dep_t_new.head()
```

### Creating a Data Dictionary

```python
# Create a data dictionary based on the revised department table
data_dict = df_dep_t_new.to_dict('index')
```
